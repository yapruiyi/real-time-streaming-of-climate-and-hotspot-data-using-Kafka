```python
# import statements
from time import sleep
from json import dumps
from kafka3 import KafkaProducer
import random
from random import randrange
from datetime import date, datetime, timedelta
from pprint import pprint
import pymongo
from pymongo import MongoClient
import pandas as pd

hostip = "192.168.0.130" # change it to your IP



def publish_message(producer_instance, topic_name, data):
    try:
        value =  bytes(str(data), 'utf-8')
        producer_instance.send(topic_name, value)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=[f'{hostip}:9092'],
                                  #value_serializer=lambda x: dumps(x, default=str).encode('utf-8'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
    
if __name__ == '__main__':
    
    client = MongoClient('192.168.0.130', 27017) 
    db = client.fit3182_db
    fit3182_assignment_db = db.fit3182_assignment_db
    
    res = fit3182_assignment_db.aggregate([
    { "$sort" : {"date" : -1 }},
    { "$limit" : 1 }
    ])
    
    
    
    # load data
    hotspot_AQUA_streaming_file = pd.read_csv("./hotspot_AQUA_streaming.csv")
    #climate_file = climate_file.apply(pd.to_numeric, errors='ignore')
    hotspot_AQUA_data = hotspot_AQUA_streaming_file.to_dict(orient="records")

    #pprint(climate_data)
    
    topic = 'Main_Stream'
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    mydate = list(res)[0]['date'] + timedelta(1)  # lastest climate date in mongoDB is 2024-01-01 so newdate should be 2024-01-02
    n = 1 # sends data every n seconds
    
    
    starttime = datetime.now()
    
    for e in range(400):
        # strftime is to format the datetime object as a string representation in HH:MM:SS.
        #time = starttime + timedelta(hours=randrange(24)) + timedelta(minutes=randrange(60)) + timedelta(seconds=randrange(60))
        time = starttime + timedelta(minutes=randrange(20)) + timedelta(seconds=randrange(60))
        data = hotspot_AQUA_data[random.randrange(0, len(hotspot_AQUA_data))]
        message = time.strftime("%H:%M:%S") + ", " + str("producer2") + " , " + str(data['latitude']) + " , "  + str(data['longitude']) + " , " + str(data['confidence']) + " / " + str(data['surface_temperature_celcius'])   
        publish_message(producer, topic, message) # " "parsed" is the key, data is the value"
        sleep(n) 
        
```
