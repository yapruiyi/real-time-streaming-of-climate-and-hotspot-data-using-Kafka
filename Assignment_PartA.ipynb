An embedded data model design is used in this assignment.

example of data model design:
 {'GHI_w/m2': 102,
  'air_temperature_celcius': 11,
  'date': Timestamp('2023-04-26 00:00:00'),
  'hotspot': [{'confidence': 56,
               'date': Timestamp('2023-04-26 00:00:00'),
               'datetime': '2023-04-26 13:26:10',
               'latitude': -36.5511,
               'longitude': 146.7819,
               'surface_temperature_celcius': 34},
               {'confidence': 82,
               'date': Timestamp('2023-04-04 00:00:00'),
               'datetime': '2023-04-04 04:32:50',
               'latitude': -37.1089,
               'longitude': 143.0783,
               'surface_temperature_celcius': 55}],
  'max_wind_speed': 20.0,
  'precipitation': '0.24G',
  'relative_humidity': 40.8,
  'station': 948701,
  'windspeed_knots': 12.2}

justification of data model design:
1) Compared to referrence data model. Related data can be stored together in embedded data model. As a result, applications require fewer queries to complete typical use cases.
2) For one climate data, it may contain zero or many fire data. This highlights a one-to-many relationship which is better suited for embedded data model. Referrence data model is better suited for many-to-many relationship.
3) Better performance for read operations compared to referrence data model, because related data can be stored together in embedded data model and there is no duplication of data for hotspots for different stations in climate data.
4) Can have the ability to update related data in a single atomic write operation. This means that only one read or write operation can be performed at a time, thus ensuring data integrity.

To complete tasks 2a and 2c, we need to filter documents by executing queries based on date matches, necessitating the creation of a date index. In task 2b, document filtering is based on queries matching hotspot.surface_temperature_celcius, hence requiring the creation of an index for hotspot.surface_temperature_celcius. Similarly, for tasks 2d and 2g, document filtering relies on queries matching hotspot.confidence, so hotspot.confidence index is created. Furthermore, for task 2j, document filtering is based on precipitation, and thus precipitation index is created. Finally, because the number of stations and data points will increase significantly once the database and application is in production, an index for station will be created as well with 
unique=True to ensure there are no duplicate stations in the document. All other indexes besides station is not set to be unique because it valid for different stations to have the same date/precipitation/confidence/surface_temperature_celcius values.
